```python
import pandas as pd
import numpy as np
```

```python
def read_csv(path_name):
    df = pd.read_csv(path_name, index_col=False)
    return df

path_name= "data.csv" 
df= read_csv(path_name)    
df.tail()
```

```python
def prop_user_sign_up(df):
    total_num_users= df.shape[0] #  number of rows
    num_sign_ups= len(df[df['171']==1])
    prop_sign_up= num_sign_ups/total_num_users
    print("number of sign up users: ", num_sign_ups)
    print("total number of users: ", total_num_users)
    print( "proportion_of_users_signed_up: ", prop_sign_up )
prop_user_sign_up(df)    
```

```python
def handle_missing(df):
    total_num_users= df.shape[0]
    
    df.drop(['172'], axis=1, inplace= True) #last col is user id
    df.rename(index=str, columns={'171': "signup"}, inplace= True) # last 2nd one is label
    
    #drop feature signup of nan
    df= df[df['signup'].notnull()]

    #drop the field(column) if more than 50% users values are missing
    cols_use= []
    for n in df.columns:
        num_of_nan= len(df[n].dropna())
        if num_of_nan> (total_num_users/2): cols_use.append(str(n))
    df= df[cols_use]
    
    #drop nan row
    df.dropna(inplace=True) 
    
    #save data
    df.to_csv('after_fix_missing.csv', index=False)
    return df
    
df= handle_missing(df)
df.head()
```
