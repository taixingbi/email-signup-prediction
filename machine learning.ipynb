```python
import pandas as pd
def read_csv(path_name):
    df = pd.read_csv(path_name, index_col=False)
    return df

path_name= "after_fix_missing.csv" 
df= read_csv(path_name)    
df.tail()
```

```python
def random_sample(df):
    #ramdon under sample not sign up
    df_sign_up= df[df['signup']==1]
    df_not_sign_up= df[df['signup']==0]

    num_sign_up= len(df_sign_up)
    df_not_sign_up= df_not_sign_up.sample(num_sign_up)
    #merge and random
    df= df_sign_up.append(df_not_sign_up)
    df= df.sample(len(df))

    print("num_sign_up: ", len(df_sign_up))
    print("num_not_sign_up: ", len(df_not_sign_up))
    
    return df

df= random_sample(df)
df.head()
```

```python
from sklearn import preprocessing
def df2array(df):
    cols = ['67', '68', '84', '85', '87', '88', '101', '102', '103', '136', '156','159']
    df_X= df[cols]
    min_max_scaler = preprocessing.MinMaxScaler()#(0, 1)
    X = min_max_scaler.fit_transform(df_X) 
    
    y= df['signup'].as_matrix()
    return X, y
X, y= df2array(df)   
print(X, y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

# KNN
#### Advantages  
1. effective if the training data is large

#### Disadvantages:
1. sensitive to noise(since it is based on the Euclidean distance), tend to overfitting 
1. need to determine value of parameter k
2. distance based learning is nor clear
3. computation cost is quite high, because need to compute distance of each query distance to all training sample


```python
from sklearn.neighbors import KNeighborsClassifier

def knn(n_neighbors):
    neigh = KNeighborsClassifier(n_neighbors=n_neighbors) #tuning parameter
    neigh.fit(X_train, y_train) 
    y_pred = neigh.predict(X_test)
    return accuracy_score(y_test, y_pred)
     
knn(17)
```

# linear sgd classifier

```python
from sklearn import linear_model
def linear_sgd():
    clf = linear_model.SGDClassifier()
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc= accuracy_score(y_test, y_pred)
    print (acc)
    
linear_sgd()
```

# logistic regression
#### Advantages:
1. Convenient probability scores for observations
2. Efficient implementations available across tools
3. Multi-collinearity is not really an issue and can be countered with L2 regularization to an extent   

#### Disadvantages:
1. Doesnâ€™t perform well when feature space is too large
2. Doesnâ€™t handle large number of categorical features/variables well
3. Relies on transformations for non-linear features


```python
from sklearn import datasets, linear_model
def logistic_regression():
    logreg = linear_model.LogisticRegression(C=1e5)
    logreg.fit(X_train, y_train)
    y_pred = logreg.predict(X_test)
    acc= logreg.score(X_test, y_test)
    print(acc)
    
logistic_regression()
```

# svm
#### Advantages  
1. Can handle large feature space
2. Can handle non-linear feature interactions

#### Disadvantages:
1. Not very efficient with large number of observations
2. It can be tricky to find appropriate kernel sometimes

```python
from sklearn import svm
def Svm():
    clf = svm.SVC()#degree: default=3
    clf.fit(X_train, y_train) 
    y_pred = clf.predict(X_test)
    acc= accuracy_score(y_test, y_pred)
    print (acc) 
Svm()
```

# Gradient Boosting regression
#### Advantages  
1. solve almost all objective function 

#### Disadvantages:
1. more sensitive to overfitting if the data is noisy. Training generally takes longer because of the fact that trees are built sequentially.
2. harder to tune than RF. There are typically three parameters: number of trees, depth of trees and learning rate, and the each tree built is generally shallow.

```python
#http://scikit-learn.org/stable/modules/ensemble.html>>> from sklearn.datasets import make_hastie_10_2
from sklearn.ensemble import GradientBoostingClassifier
def graboosting():
    clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
          max_depth=1, random_state=0).fit(X_train, y_train)
    acc= clf.score(X_test, y_test) 
    print(acc)
graboosting()    
```
